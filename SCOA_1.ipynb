**Union of Fuzzy sets**

````python
A , B , Y = dict() , dict() , dict()

A = {"x1": 0.1, "x2": 0.7, "x3": 1, "x4": 0}
B = {"x1": 0.4, "x2": 0.3, "x3": 0.2, "x4": 1}

print('The First Fuzzy Set is :', A)
print('The Second Fuzzy Set is :', B)

for A_key, B_key in zip(A, B):
	A_value = A[A_key]
	B_value = B[B_key]

	if A_value > B_value:
		Y[A_key] = A_value
	else:
		Y[B_key] = B_value

print('Fuzzy Set Union is :', Y)
````
